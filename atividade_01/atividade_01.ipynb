10.  Utilizando uma query em SQL e o banco `sqlite-sakila.db` retorne as cidades e os países, mas apenas os países que começem com a letra **B**.

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
engine = create_engine('sqlite:///sqlite-sakila.db')
insp = inspect(engine)
```

```python
insp.get_table_names()
```

```python
sql_query = '''
    select
        *
    from 
        country
    where
        country like "B%"
'''
countries_with_b = pd.read_sql_query(sql_query, engine)
countries_with_b
```
